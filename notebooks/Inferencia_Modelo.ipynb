```python
# Función para convertir de avi to mp4 para visualizar en Colab
def convertir_avi_to_mp4(avi_file_path, output_name):
    os.popen("ffmpeg -i '{input}' -ac 2 -b:v 2000k -c:a aac -c:v libx264 -b:a 160k -vprofile high -bf 0 -strict experimental -f mp4 '{output}.mp4'".format(input = avi_file_path, output = output_name))
    return True

def esperar_elemento(nombre_elemento, carpeta, tiempo_maximo_espera=60):
    tiempo_inicio = time.time()

    while not os.path.exists(os.path.join(carpeta, f'{nombre_elemento}.mp4')):
        tiempo_actual = time.time()
        tiempo_transcurrido = tiempo_actual - tiempo_inicio

        if tiempo_transcurrido >= tiempo_maximo_espera:
            print(f"El elemento {nombre_elemento} no se encontró en {tiempo_maximo_espera} segundos.")
            break

        time.sleep(1)

    if os.path.exists(os.path.join(carpeta, nombre_elemento)):
        print(f"El elemento '{nombre_elemento}.mp4' se encontró en {tiempo_transcurrido:.2f} segundos.")
```
